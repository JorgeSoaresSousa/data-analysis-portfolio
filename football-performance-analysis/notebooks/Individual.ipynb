```python
import pandas as pd

df_individual = pd.read_csv("Individual.csv")
df_individual.head()
```

```python

df_individual["Min"] = df_individual["Min"].replace(0, pd.NA)

df_individual["Desarmes_per90"] = df_individual["TklW"] / df_individual["Min"] * 90
df_individual["Bloqueios_per90"] = df_individual["Blocks"] / df_individual["Min"] * 90
df_individual["Intercessoes_per90"] = df_individual["Int"] / df_individual["Min"] * 90
df_individual["BolaAerea_per90"] = df_individual["AerialsWon"] / df_individual["Min"] * 90
df_individual["ConduçaoPrgc_per90"] = df_individual["PrgC"] / df_individual["Min"] * 90
```

```python
cols_keep = [
    "Player",
    "Pos",
    "PassesCmp",
    "PassesCmp%",
    "Att",
    "Succ%",
    "Sh",
    "G/Sh",
    "SoT%",
    "SCA90",
    "Desarmes_per90",
    "Bloqueios_per90",
    "Intercessoes_per90",
    "BolaAerea_per90",
    "ConduçaoPrgc_per90"
]

df_individual = df_individual[cols_keep]
df_individual.head()
```

```python
df_individual = df_individual.copy()

numeric_cols = [
    "PassesCmp",
    "PassesCmp%",
    "Att",
    "Succ%",
    "Sh",
    "G/Sh",
    "SoT%",
    "SCA90",
    "Desarmes_per90",
    "Bloqueios_per90",
    "Intercessoes_per90",
    "BolaAerea_per90",
    "ConduçaoPrgc_per90"
]

for col in numeric_cols:
    df_individual[col] = (
        df_individual[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .astype(float)
    )


```

```python
output_path = r"C:\Users\jjorg\Desktop\Analise Futebol\data_processed\individual_cleaned.csv"
df_individual.to_csv(output_path, index=False, encoding="utf-8")
```
