```python
import pandas as pd

df = pd.read_csv("top5_stats.csv")
df.head()
```

```python
df["eff_ofensiva"] = df["Gls"] / df["xG"]
df["eff_defensiva"] = df["xGA"] / df["GlsA"]
df["prg_total"] = df["PrgC"] + df["PrgP"]
df["g_per_90"] = df["Gls"] / df["90s"]
df["ga_per_90"] = df["GlsA"] / df["90s"]
df["xg_per_90"] = df["xG"] / df["90s"]
df["xga_per_90"] = df["xGA"] / df["90s"]
```

```python
df_overview = df[['Squad', 'Poss', 'Gls', 'xG', 'GlsA', 'xGA']]  
df_overview.to_csv(r"C:\Users\jjorg\Desktop\Analise Futebol\overview_top5.csv", index=False)
```
