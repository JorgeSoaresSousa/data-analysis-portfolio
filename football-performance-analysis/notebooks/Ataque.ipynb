```python
import pandas as pd

df_attack = pd.read_csv("ataque.csv")
```

```python
df_attack = df_attack.drop(columns=["Sh", "Sh.1"])
df_attack = df_attack.rename(columns={"Sh.2": "Shots"})
```

```python
df_attack["SoT"] = 0
df_attack["SoT"] = [
    38, 33, 34, 64, 56, 33, 47, 40, 48, 49,
    38, 39, 66, 44, 41, 82, 43, 41
]
```

```python
df_attack[["Squad", "Shots", "SoT"]]
```

```python

df_attack["ShotPr"] = (df_attack["SoT"] / df_attack["Shots"]) * 100

df_attack["Finalizacao"] = (df_attack["Gls"] / df_attack["Shots"]) * 100

df_attack["ShotPr"] = df_attack["ShotPr"].round(2)
df_attack["Finalizacao"] = df_attack["Finalizacao"].round(2)



```

```python
df_attack.head()
```

```python
df_attack = df_attack[[
    "Squad",
    "Finalizacao",
    "ShotPr",
    "SCA",
    "PassLive",
    "PassDead"
]].copy()
equipas_top = [
    "Famalicão",
    "Benfica",
    "Gil Vicente FC",
    "Sporting CP",
    "Porto"
]

df_attack = df_attack[df_attack["Squad"].isin(equipas_top)].reset_index(drop=True)

```
